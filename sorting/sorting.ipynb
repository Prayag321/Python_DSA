			Selection Sort
	</h1>
<marquee>

```python
def selection_sort(arr):
  """
  Description:
  	Perform selection sort on a list.

  Parameters:
  	arr (list): The list of elements to be sorted.

  Returns:
  	list: The sorted list.
  """
  # Traverse through all elements in the list
  for i in range(len(arr)):
    # Find the minimum element in the remaining unsorted list
    min_idx = i
    for j in range(i + 1, len(arr)):
      if arr[j] < arr[min_idx]:
        min_idx = j

    # Swap the values if small element found
    arr[i], arr[min_idx] = arr[min_idx], arr[i]

  return arr

if __name__ == "__main__":
  sample_list = [64, 25, 12, 22, 11]
  print("Original list:", sample_list)
  sorted_list = selection_sort(sample_list)
  print("Sorted list:", sorted_list)
```

			Bubble Sort

<marquee>

```python
def bubble_sort(arr):
  """
  Description:
    Perform bubble sort on a list.

  Parameters:
    arr (list): The list of elements to be sorted.

  Returns:
    list: The sorted list.
  """
  n = len(arr)
  for i in range(n):
    # Flag to detect if any swapping happened
    swapped = False
    # Traverse through 0 to n-i-1
    for j in range(0, n-i-1):
      # Swap if the element found is greater than the next element
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        swapped = True
    # If no swapping happened, the array is already sorted
    if not swapped:
      break
  return arr

def main():
  arr = [64, 34, 25, 12, 22, 11, 90]
  print("Original array:", arr)
  sorted_arr = bubble_sort(arr)
  print("Sorted array:", sorted_arr)

if __name__ == "__main__":
  main()
```

```python
def merge_sort(arr):
  """
  Description:
    Perform merge sort on a list.

  Parameters:
    arr (list): The list of elements to be sorted.

  Returns:
    list: The sorted list.
  """
  if len(arr) > 1:
    # Find the middle of the list
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursively sort both halves
    merge_sort(left_half)
    merge_sort(right_half)

    # Merge sorted halves
    i = j = k = 0

    while i < len(left_half) and j < len(right_half):
      if left_half[i] < right_half[j]:
        arr[k] = left_half[i]
        i += 1
      else:
        arr[k] = right_half[j]
        j += 1
      k += 1

    # Copy the remaining elements of left_half, if any
    while i < len(left_half):
      arr[k] = left_half[i]
      i += 1
      k += 1

    # Copy the remaining elements of right_half, if any
    while j < len(right_half):
      arr[k] = right_half[j]
      j += 1
      k += 1

  return arr

def main():
  arr = [64, 34, 25, 12, 22, 11, 90]
  print("Original array:", arr)
  sorted_arr = merge_sort(arr)
  print("Sorted array:", sorted_arr)

if __name__ == "__main__":
  main()
```
